### Steps to building a neural network using `tensorflow.keras`

1. `my_model=Sequential()` Creates empty model (neural network) 
2. `my_model.add(Dense(16), input_shape=(14,), activation='relu')` adds a layer to model with 16 (arbitrary number) nodes and ReLU activation function; repeat this several times (without input_shape) to add more layers
3. `my_model.compile(optimizer='adam', loss='mse')`: config the model with optimizer and loss function
4. `my_model.fit(X_train, y_train, batch_size=32, epochs=10)` train the model
5. `my_model.predict(X_test)` to run inference/predictions
6. (optional) `model.evaluate(X_test, y_test)` to evaluate the model perfomance on X_test

```python
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
# Make up some random data

X=np.linspace(0,10,500)
y=np.sin(X)
```

```python
X, y
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.scatter(X, y)
```

```python
train_test_split?
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train
```

```python
model=Sequential() # Creates empty model
model.add(Dense(10, input_shape=(1,)))  #adds a layer with 10 nodes, no activation function
```

```python
model.summary()
```

```python
model.add(Dense(10))
model.add(Dense(1))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
model.evaluate(X_test, y_test)
```

```python
fig=plt.figure()
ax=fig.add_subplot()
ax.scatter(X_test, y_test)
ax.scatter(X_test, model.predict(X_test), color='red')
```

```python
for i in range(10):
    model.fit(X_train, y_train, epochs=10)
    fig=plt.figure()
    ax=fig.add_subplot()
    ax.scatter(X_test, y_test)
    ax.scatter(X_test, model.predict(X_test), color='red')
    fig.show()
    print('Loss on test set:', model.evaluate(X_test, y_test))
```
